# Classification Decision Tree

A **Classification Decision Tree** addresses problems where the goal is to predict a **categorical outcome** based on input features. This method creates a tree-like model of decisions, where:

- Each **internal node** represents a test on an attribute.
- Each **branch** represents the outcome of the test.
- Each **leaf node** represents a class label or category.

## Problems Addressed by a Classification Decision Tree
These are typically **classification problems**, where the output is a discrete label or category. Examples include:
- **Binary classification**: Two possible outcomes (e.g., "yes" or "no").
- **Multi-class classification**: More than two categories (e.g., "low," "medium," "high").

## Real-World Applications

1. **Medical Diagnosis**  
   - Classifying a disease as present or absent based on symptoms and test results.  
   - Example: Determining whether a patient has diabetes based on glucose levels and age.

2. **Credit Risk Assessment**  
   - Categorizing loan applicants into "low risk" or "high risk" based on their credit history and income.

3. **Customer Segmentation**  
   - Grouping customers into categories such as "frequent shopper" or "one-time buyer" based on their purchasing behavior.

4. **Fraud Detection**  
   - Identifying whether a credit card transaction is "fraudulent" or "legitimate" based on patterns in transaction data.

5. **Spam Email Detection**  
   - Determining whether an incoming email is "spam" or "not spam" based on keywords, sender information, and message structure.

6. **Predictive Maintenance**  
   - Classifying machinery as "functional," "needs maintenance soon," or "non-functional" based on sensor data.

---

# Difference Between Classification Decision Tree and Multiple Linear Regression

## Classification Decision Tree
- **Purpose:** Used for **classification tasks**, where the target variable is categorical (e.g., "yes" or "no," "class A" or "class B").
- **How It Predicts:**
  1. The tree splits data into subsets based on feature values at each node.
  2. At each split, it chooses the feature that best separates the classes (using metrics like **Gini impurity** or **entropy**).
  3. At the terminal nodes (leaves), the majority class (or class probabilities) is used to make the prediction.
- **Output:** Class label or class probabilities.

### Example:
If you want to predict whether a student will **pass** or **fail** based on their study hours and attendance, the tree might create rules like:
- If study hours > 5 and attendance > 80%, predict "pass."
- Otherwise, predict "fail."

---

## Multiple Linear Regression
- **Purpose:** Used for **regression tasks**, where the target variable is continuous (e.g., house prices, temperatures).
- **How It Predicts:**
  1. Models the relationship between the target variable and input features as a linear equation:  
     \[
     y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \ldots + \epsilon
     \]
  2. Coefficients (\(\beta\)) represent the contribution of each feature to the outcome.
  3. Predictions are made by substituting the feature values into the equation.
- **Output:** A numeric value.

### Example:
If you want to predict a house's price based on its size and number of bedrooms, the model might predict:
\[
\text{price} = 50,000 + 100 \cdot \text{size} + 20 \cdot \text{bedrooms}
\]

---

## Key Differences

| Feature                  | Classification Decision Tree                  | Multiple Linear Regression               |
|--------------------------|-----------------------------------------------|------------------------------------------|
| **Target Variable**      | Categorical                                  | Continuous                               |
| **Output**               | Class label or probabilities                 | Numeric value                            |
| **Model**                | Decision rules based on feature splits       | Linear equation                          |
| **Type of Prediction**   | Classifies into discrete categories          | Predicts a continuous outcome            |
| **Interpretation**        | Branches and splits                         | Coefficients indicating feature impacts  |

# Real-World Applications for Evaluation Metrics

## 1. **Accuracy**
- **Formula:**  
  \[
  \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
  \]
- **Scenario:** **General-purpose classification with balanced classes**  
  Example: Predicting whether an email is spam or not in a balanced dataset.  
  - **Rationale:** Accuracy is effective when false positives (FP) and false negatives (FN) have similar costs and the dataset is balanced (equal number of positive and negative cases).

---

## 2. **Sensitivity (Recall or True Positive Rate)**
- **Formula:**  
  \[
  \text{Sensitivity} = \frac{TP}{TP + FN}
  \]
- **Scenario:** **Medical diagnosis**  
  Example: Detecting cancer in patients based on medical tests.  
  - **Rationale:** Sensitivity ensures the model captures most of the actual positive cases (e.g., true cancer cases). Missing a positive case (FN) could have severe consequences for a patient's health.

---

## 3. **Specificity**
- **Formula:**  
  \[
  \text{Specificity} = \frac{TN}{TN + FP}
  \]
- **Scenario:** **Screening for rare conditions**  
  Example: A test to rule out a disease in a population where the disease is rare.  
  - **Rationale:** Specificity ensures that most of the actual negatives (healthy individuals) are correctly identified. This reduces unnecessary worry and further testing caused by false positives (FP).

---

## 4. **Precision (Positive Predictive Value)**
- **Formula:**  
  \[
  \text{Precision} = \frac{TP}{TP + FP}
  \]
- **Scenario:** **Fraud detection in banking**  
  Example: Identifying fraudulent transactions from legitimate ones.  
  - **Rationale:** Precision ensures that flagged transactions (positives) are highly likely to be fraudulent. Reducing false positives (FP) is crucial to avoid inconveniencing legitimate customers.

---

### Summary Table

| Metric         | Best Application Scenario                      | Rationale                                                                  |
|----------------|-------------------------------------------------|---------------------------------------------------------------------------|
| **Accuracy**   | Balanced datasets with equal costs for FP & FN | Effective when class distributions are even and costs are balanced.       |
| **Sensitivity**| High-risk detection (e.g., medical tests)      | Ensures most actual positives are identified, reducing false negatives.   |
| **Specificity**| Rare condition exclusion                       | Ensures most actual negatives are identified, reducing false positives.   |
| **Precision**  | High-confidence positive predictions           | Ensures that positive predictions are highly accurate, reducing FP costs. |

In the classification tree created by clf2, predictions are made based on the following general process:


Root node: The first split will be based on one of the features (NumPages, Thick, List Price). This split will be made by choosing the feature and value that best divides the data into the two classes (Hardcover vs. Paperback). The algorithm uses criteria like Gini impurity or entropy to determine the best split.


Internal nodes: Each subsequent split will follow the same logic, splitting based on the next best feature at that node. The decision will be made recursively down the tree.


Leaf nodes: Once the tree reaches the maximum depth (4 in this case), the leaf nodes represent the final predicted class. These are the predicted values for whether a book is hardcover (1) or paperback (0). The leaf node will show the class prediction along with the proportion of books in each class for that node.


The tree uses the following features to make predictions:


NumPages: The number of pages in the book. For instance, thicker books (with more pages) might be more likely to be hardcover.

Thick: This is a measure of the thickness of the book. If a book is thick, it might influence whether it’s hardcover.

List Price: Higher-priced books may tend to be hardcover, as hardcover books are typically more expensive.
The decision tree will use the combinations of these features to split the dataset into smaller groups at each level, ultimately predicting whether a book is a hardcover or paperback.
f

How Predictions Are Made:
Given new data (e.g., a book with specific NumPages, Thick, and List Price), the model will traverse the tree starting at the root node, checking which condition holds (e.g., NumPages <= 350). It will move down through the tree, making splits based on the other features, until it reaches a leaf node.

The final prediction will be the class (Hardcover or Paperback) corresponding to the majority class in that leaf node.

````python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Prepare the test set target and features for both models
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])["H"]
X_test_clf = ab_reduced_noNaN_test[["List Price"]]
X_test_clf2 = ab_reduced_noNaN_test.copy()

# Simulating the missing test features for clf2
X_test_clf2['NumPages'] = [250, 120]  # Example values
X_test_clf2['Thick'] = [1.4, 0.9]
X_test_clf2 = X_test_clf2[["NumPages", "Thick", "List Price"]]

# Predictions for clf and clf2
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrices
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Sensitivity (Recall), Specificity, Accuracy
def calculate_metrics(cm):
    TN, FP, FN, TP = cm.ravel()
    sensitivity = TP / (TP + FN) if (TP + FN) > 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    return round(sensitivity, 3), round(specificity, 3), round(accuracy, 3)

# Metrics for clf and clf2
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(cm_clf)
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(cm_clf2)

# Display confusion matrices
disp_clf = ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=["Paper Back", "Hard Cover"])
disp_clf.plot(cmap="Blues", values_format=".3f")

disp_clf2 = ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=["Paper Back", "Hard Cover"])
disp_clf2.plot(cmap="Blues", values_format=".3f")

# Print metrics
print("Metrics for clf (List Price only):")
print(f"Sensitivity: {sensitivity_clf}, Specificity: {specificity_clf}, Accuracy: {accuracy_clf}\n")

print("Metrics for clf2 (NumPages, Thick, List Price):")
print(f"Sensitivity: {sensitivity_clf2}, Specificity: {specificity_clf2}, Accuracy: {accuracy_clf2}")
````

# Reporting Sensitivity, Specificity, and Accuracy

---

## Definitions:
1. **Sensitivity (Recall):** Measures how well the model identifies positive cases.
   \[
   \text{Sensitivity} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Negatives (FN)}}
   \]

2. **Specificity:** Measures how well the model identifies negative cases.
   \[
   \text{Specificity} = \frac{\text{True Negatives (TN)}}{\text{True Negatives (TN)} + \text{False Positives (FP)}}
   \]

3. **Accuracy:** Measures the overall correctness of the model.
   \[
   \text{Accuracy} = \frac{\text{True Positives (TP)} + \text{True Negatives (TN)}}{\text{Total Observations}}
   \]

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does.

But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. ## 1. Understanding Classification Decision Trees

### (a) What problem does a Classification Decision Tree address, and where is it applied?
A Classification Decision Tree addresses problems involving categorical outcomes. It creates a tree-like structure where:
- **Nodes** represent tests on features.
- **Branches** represent outcomes.
- **Leaves** represent class labels.

#### Examples of applications:
- **Medical Diagnosis**: Classifying diseases as "present" or "absent."
- **Credit Risk Assessment**: Categorizing loan applicants as "low risk" or "high risk."
- **Fraud Detection**: Flagging fraudulent vs. legitimate transactions.
- **Spam Detection**: Classifying emails as "spam" or "not spam."

### (b) Differences between Classification Decision Trees and Multiple Linear Regression
| Feature                  | Classification Decision Tree            | Multiple Linear Regression     |
|--------------------------|-----------------------------------------|--------------------------------|
| **Target Variable**      | Predicts categories                    | Predicts continuous outcomes  |
| **Model Structure**      | Decision rules splitting data recursively | Linear equations with coefficients |
| **Output**               | Class labels or probabilities          | Numeric values                |

---

## 2. Real-World Applications for Evaluation Metrics

- **Accuracy**: Best for balanced datasets with equal costs for FP & FN (e.g., spam detection).
- **Sensitivity**: Critical in high-risk applications like medical diagnosis, where missing positives is costly.
- **Specificity**: Useful for screening rare conditions to reduce false positives.
- **Precision**: Important in scenarios like fraud detection, where positive identifications need high confidence.
